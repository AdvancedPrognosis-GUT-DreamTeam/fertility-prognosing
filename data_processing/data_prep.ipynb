```python
import pandas as pd
```

```python
df = pd.read_csv('wdi_fertility_data.csv')
df.head(10)
```

```python
df.columns
```

```python
COLUMN_NAME_MAP = {
    "Time": "year",
    "Time Code": "year_code",
    "Country Name": "country_name",
    "Country Code": "country_code",
    "Fertility rate, total (births per woman) [SP.DYN.TFRT.IN]": "total_fertility_rate",
    "GDP (constant 2015 US$) [NY.GDP.MKTP.KD]": "gdp_2015_dollar",
    "GDP per capita (constant 2015 US$) [NY.GDP.PCAP.KD]": "gdp_per_capita_2015_dollar",
    "GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]": "gdp_growth",
    "Physicians (per 1,000 people) [SH.MED.PHYS.ZS]": "physicians_per_1000",
    "Hospital beds (per 1,000 people) [SH.MED.BEDS.ZS]": "hospital_beds_per_1000",
    "Arable land (hectares per person) [AG.LND.ARBL.HA.PC]": "arable_land_hectare_per_person",
    "Consumer price index (2010 = 100) [FP.CPI.TOTL]": "cpi",
    "Crop production index (2014-2016 = 100) [AG.PRD.CROP.XD]": "crop_production_index",
    "Food production index (2014-2016 = 100) [AG.PRD.FOOD.XD]": "food_production_index",
    "Households and NPISHs final consumption expenditure (% of GDP) [NE.CON.PRVT.ZS]": "percent_of_household_expenditure_in_gdp",
    "Life expectancy at birth, total (years) [SP.DYN.LE00.IN]": "life_expectancy_at_birth",
    "Mortality rate, infant (per 1,000 live births) [SP.DYN.IMRT.IN]": "infant_mortality_rate",
    "Mortality rate, under-5 (per 1,000 live births) [SH.DYN.MORT]": "child_mortality_rate",
    "Educational attainment, at least Bachelor's or equivalent, population 25+, total (%) (cumulative) [SE.TER.CUAT.BA.ZS]": "percentage_of_bachelor_degree",
    "Labor force participation rate, female (% of female population ages 15+) (national estimate) [SL.TLF.CACT.FE.NE.ZS]": "female_percentage_in_labor",
    "Children in employment, total (% of children ages 7-14) [SL.TLF.0714.ZS]": "percentage_of_employed_children"
}
```

```python
df = df.rename(columns=COLUMN_NAME_MAP)
df.head(10)
```

```python
df = df[df["total_fertility_rate"].notna()]
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.isna().groupby(df['country_name']).sum()
```

```python
df.isna().groupby(df['year']).sum()
```

```python
df = df.drop(columns=["percentage_of_employed_children", "percentage_of_bachelor_degree"])
```

```python
df = df[(df['year'] >= 1990) & (df['year'] <= 2021)] 
```

```python
df.isna().sum()
```

```python
prepped_df = df
prepped_df.to_csv("wdi_fertility_data_prepped.csv")
```
